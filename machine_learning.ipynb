```python
# vamos a visualizar la relacion que hay entre la longitud y el ancho de los sepalos llevando ambas
# caracteristicas a un diagrama de 
fig, ax = plt.subplots(figsize = (7, 7))
ax.set_aspect("equal")
LstedColormap = ["#E67332", "#FADB15", "#4193E5"]
scatter = ax.scatter(
    x = iris["sepal_length"], y = iris["sepal_width"],
    c = iris.label, zorder = 2, edgecolor = "#999999",
)
ax.set_title("Sepal length vs Sepal width")
ax.set_xlabel("Sepal length (cm)")
ax.set_ylabel("Sepal width (cm) ")
ax.legend(
    handles = scatter.legend_elements()[0],
    labels = list(iris.species.unique())
)
ax.grid(color = "#EEEEEE", zorder = 1, alpha = 0.9)
plt.show    
```
