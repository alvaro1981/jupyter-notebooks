```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)
print(t)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r*', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a' , 'b', c='c', s='d', data= data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]
plt.plot(steps_walked)
```

```python

plt.plot(days, steps_walked, "x:r")
#adding titles and labels
plt.title("Step count for the week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked")
plt.show()

```

```python
#now add a second list with the number of steps walked the previous week so that you can compare this week’s step count with the previous week’s:
steps_last_week = [9788, 8710, 5308, 17630, 21309, 4002, 5223]
plt.plot(days, steps_walked, "o-g")
plt.plot(days, steps_last_week, "v--m")
plt.title("Step count | This week and last week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked")
plt.legend(["This week", "Last week"])
plt.grid(True)
```

```python
""" recreate the last figure you plotted earlier using the object-oriented method. To create a figure, 
you can use the function subplots(), which returns a tuple containing a Figure object and an Axes 
object when it’s called without arguments """

fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.plot(days, steps_walked, "o-g")
ax.plot(days, steps_last_week, "v--m")
ax.set_title("Step count | This week and last week")
ax.set_xlabel("Days of the week")
ax.set_ylabel("Steps walked")
ax.grid(True)
ax.legend(["This week", "Last week"])
```
