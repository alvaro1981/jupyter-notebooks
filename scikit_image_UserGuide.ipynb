```python
import skimage as ski
import numpy as np
import matplotlib.pyplot as plt
```

```python
camera = ski.data.camera()
```

```python
plt.imshow(camera)
```

```python
# Get the vale of the pixel at 10th row and 20th column
camera[10, 20]
# Set to black the pixel at the 3rd row and 10th column
camera[3, 10] = 0
```

```python
# Set the first ten lines to "black" (0)
camera[:10] = 0
plt.imshow(camera)
camera[:11]
```

```python
mask = camera < 87
# Set to "white" (255) the pixels where mask is True
camera[mask] = 255
plt.imshow(camera)
camera [:12]
```

```python
#print(camera)
print(len(camera))
inds_r = np.arange(len(camera))

#print(inds_r)
inds_c = 4 * inds_r % len(camera)
#print(inds_c)

camera[inds_r, inds_c] = 0
plt.imshow(camera)


```

```python
nrows, ncols = camera.shape
row, col = np.ogrid[:nrows, :ncols]
print(f"ROW shape  = {row.shape}  COL shape = {col.shape}")
#print("ROW : " ,row)
#print("COL : " ,col)
cnt_row, cnt_col = nrows / 2, ncols / 2
outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 >
                   (nrows / 2)**2)
print(outer_disk_mask)
camera[outer_disk_mask] = 0
plt.imshow(camera)
```

```python
lower_half = row > cnt_row
#print(lower_half)
lower_half_disk = np.logical_and(lower_half, outer_disk_mask)
camera = ski.data.camera()
camera[lower_half_disk] = 0
plt.imshow(camera)
```
