```python
import matplotlib.pyplot as plt
import numpy as np

wavelength = 1
x = np.linspace(-1, 1, 1000)
y = np.sin(2 * np.pi * x / wavelength)
print(f" type(y) = {type(y)}, y.shape = {y.shape}")
print(f" type(x) = {type(x)}, x.shape = {x.shape}")
plt.plot(x, y)
```

```python
x = y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x,y)
print(f" type(X) = {type(X)} , X.shape = {X.shape}")
wavelength=1
output = np.sin(2 * np.pi * X / wavelength)
plt.imshow(output)
```

```python
output = np.sin(2 * np.pi * Y / wavelength)
plt.imshow(output)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': "3d"})
ax.plot_surface(X, Y, output, cmap="viridis")
```

```python
wavelength = 0.5
x = y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)
output = np.exp(-5 * (X ** 2) / 2 - 5 * (Y ** 2) / 2)
fig = plt.figure()

ax = fig.add_subplot(121)
ax.imshow(
    output,
    cmap="copper",
    extent=[np.min(x), np.max(x), np.min(y), np.max(y)],
)
ax = fig.add_subplot(122, projection="3d")
ax.plot_surface(X, Y, output, cmap="copper")

```
