```python
import pandas as pd

df = pd.read_csv('nba.csv')
df.head()

```

```python
team = df.groupby('Team')
print(type(team))
team.first()
```

```python
grouping = df.groupby(['Team','Position'])
grouping.first()
```

```python
# Applying Aggregation with GroupBy
aggregated_data  = df.groupby(['Team','Position']).agg(
    total_salary=('Salary', 'sum'),
    avg_salary=('Salary', 'mean'),
    player_count=('Name', 'count')
)
print(type(aggregated_data))
aggregated_data
```

```python
# Selecciono el rango de equipos que quiero ver del dataFrame aggregate_data
selected_rows = aggregated_data.loc['Chicago Bulls':'Houston Rockets'] 	
selected_rows
```

```python
# Rank players within each team by Salary
df['Rank within Team'] = df.groupby('Team')['Salary'].transform(lambda x:
x.rank(ascending=False))
df
```

```python
#Filter groups where the average Salary is >= 6 million
filtered_df = df.groupby('Team').filter(lambda x:x['Salary'].mean() >= 6000000)
filtered_df
```

```python
#Filter groups where the min Salary is  900000
filtered_df = df.groupby('Team').filter(lambda x:x['Salary'].min() >= 900000)
filtered_df
```

```python
  
# making data frame from csv file 
data = pd.read_csv("nba.csv") 
  
# creating a rank column and passing the returned rank series 
data["Rank"] = data["Name"].rank() 
  
# display 
data 
  
```

```python
# sorting w.r.t name column 
data.sort_values("Name", inplace = True) 
# display after sorting w.r.t Name column 
data 
```

```python
#sorting Team name
data.sort_values('Team', inplace=True)
data
```

```python
# creating a rank column and passing the returned rank series 
# change method to 'min' to rank by minimum 
data["Rank"] = data["Team"].rank(method ='average') 
data
```
