```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
X, Y = np.meshgrid(x, y)
print(X.shape)
# Temporary code: Visualising X and Y returned by meshgrid()
fig, axs = plt.subplots(1, 2)
axs[0].imshow(X, cmap="gray")
axs[0].set_title("X")
axs[1].imshow(Y, cmap="gray")
axs[1].set_title("Y")
```

```python
Z = (np.sin(X)/X) * (np.sin(Y)/Y)
plt.imshow(Z)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': "3d"})
ax.plot_surface(X, Y, Z, cmap="viridis")
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)
fig, ax = plt.subplots()
a = 0
Z = np.sin(X ** 2 + Y ** 2 - a)
ax.imshow(Z)
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)
fig, ax = plt.subplots()
for a in np.linspace(-np.pi, np.pi, 50):
    Z = np.sin(X ** 2 + Y ** 2 - a)
    ax.imshow(Z)
    plt.pause(1)
plt.show()
```

```python
X, Y = np.meshgrid(x, y)
fig, ax = plt.subplots()
file_number = 0
for a in np.linspace(-np.pi, np.pi, 50):
    Z = np.sin(X ** 2 + Y ** 2 - a)
    ax.imshow(Z)
    print(f"Saving image {file_number + 1}")
    fig.savefig(f"image_{file_number}.png")
    file_number += 1
```

```python
import matplotlib.animation as animation
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)
fig, ax = plt.subplots()
images = []
for a in np.linspace(-np.pi, np.pi, 50):
    Z = np.sin(X ** 2 + Y ** 2 - a)
    img = ax.imshow(Z)
    images.append([img])
output = animation.ArtistAnimation(fig, images, interval=50, blit=True)
plt.show()
```
