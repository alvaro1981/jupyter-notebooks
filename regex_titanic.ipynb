```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv('titanic_train.csv')
df
```

```python
expression = 'William'
name_search = re.search(expression, "Allen, Mr. William Henry")
print(name_search)
if name_search:
    print(name_search.group())

```

```python
expression = '[Ww]illiam'  # Cualquiera W o w, seguido de "illiam"
name_search = re.search(expression, "Allen, Mr. william Henry")
if name_search:
    print(name_search.group())
    
```

```python
expression = '(?i)william'          # case insensitive search
name_search = re.search(expression, "Allen, Mr. WILLIAM Henry")
if name_search:
    print(name_search.group())    
    # WILLIAM
```

```python
expression = 'Henry$'               # ends with Henry 
name_search = re.search(expression, "Allen, Mr. William Henry")
if name_search:
    print(name_search.group())
    # Henry
```

```python
expression = '^Allen'               # starts with Allen
name_search = re.search(expression, "Allen, Mr. William Henry")
if name_search:
    print(name_search.group())
    # Allen
else:
    print("La cadena no tiene el patron especificado al comienzo de la frase, retorno : ", name_search())
    
    
```

```python
expression = '^Allen'               # starts with Allen
name_search = re.search(expression, "Alvaro, Mr. William Henry")
if name_search:
    print(name_search.group())
    # Allen
else:
    print("La cadena no tiene el patron especificado al comienzo de la frase, retorno : ", name_search)
```

```python
expression = ' ([A-Za-z]+).'
name_search = re.search(expression, "Allen, Mr. William Henry")
print(name_search)
if name_search:
    print(name_search.group(0))
    print(name_search.group(1))    
    
```

```python
expression = ' ([A-Za-z]+). ([A-Za-z]+).'
titles_search = re.search(expression, "Allen, Dr. Mr. William Henry")
if titles_search:
    print(titles_search.group(0)) # Dr. Mr.
    print(titles_search.group(1)) # Dr
    print(titles_search.group(2)) # Mr
    
```

```python
# returns a list of non-overlapping matches
titles_search = re.findall('([A-Za-z]+).', "Mrs. Mr. Ms. Dr.")
print(titles_search)
```
